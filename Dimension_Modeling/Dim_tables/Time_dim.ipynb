```
from pyspark.sql.functions import *
```

```
startHour = 1
endHour = 24

df = spark.range(startHour, endHour +1, step = 1).toDF("timeID")
```

```
display(df)
```

```
df = df.withColumn("startHour", expr("current_date() + (timeID - 1) * interval 1 hour"))

df = df.withColumn("endHour", expr("startHour + interval 59 minutes 59 seconds"))

df = df.withColumn("startHour", date_format("startHour", "HH:mm:ss"))
df = df.withColumn("endHour", date_format("endHour", "HH:mm:ss"))
```

```
df.show(25)
```

```
df.write.format("delta").mode("overwrite").saveAsTable("time_dim")
```

```
display(spark.sql("select * from time_dim order by timeID"))
```
