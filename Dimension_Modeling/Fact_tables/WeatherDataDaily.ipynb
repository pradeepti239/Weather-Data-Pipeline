```
from pyspark.sql import functions as F
from pyspark.sql.functions import col,avg, min, max, first
from pyspark.sql.types import IntegerType, DoubleType, FloatType, StructType, StructField
```

```
weather_fact_hourly_df = spark.read.parquet("dbfs:/tmp/weather_fact_hourly.parquet")

display(weather_fact_hourly_df)
```

```
weather_fact_daily_df = weather_fact_hourly_df.groupBy("dateID", "cityID").agg(
    avg(col("weather_temp")).alias("avg_weather_temp"),
    avg(col("temp")).alias("avg_temp"),
    avg(col("feels_like")).alias("avg_feels_like"),
    min(col("temp_min")).alias("min_temp_min"),
    max(col("temp_max")).alias("max_temp_max"),
    avg(col("pressure")).alias("avg_pressure"),
    avg(col("humidity")).alias("avg_humidity"),
    first(col("sea_level")).alias("first_sea_level"),
    avg(col("visibility")).alias("avg_visibility"),
    avg(col("wind_speed")).alias("avg_wind_speed"),
    avg(col("clouds_all")).alias("avg_clouds_all")
)
```

```
display(weather_fact_daily_df)
```

```
weather_fact_daily_df.printSchema()
```

```
weather_fact_daily_df.write.mode("append").parquet("/tmp/weather_fact_daily.parquet")
```
